```python
from datasets import load_dataset, DatasetDict
import pandas as pd
from pathlib import Path
from glob import glob
short_name = "open-instruct-inference"
dataset_path = f'./dataset/{short_name}/{short_name}-output/'

print(glob(Path(dataset_path).as_posix() + "/*"))

dataset = DatasetDict.load_from_disk(Path(dataset_path).as_posix())
df = pd.DataFrame(dataset["predictions"])
```

```python
for index, row in df.iterrows():
    print("Prompt:\n--------------------------------------------------------------")
    print(row["prompt"])
    print("--------------------------------------------------------------\n\nActual:\n--------------------------------------------------------------")
    print(row["actual"])
    print("--------------------------------------------------------------\n\nPredicted:\n--------------------------------------------------------------")
    print(row["predicted"])

    print("\n\n")
    break
```

```python
from IPython.display import display, HTML
pd.set_option('display.max_colwidth', 50) # Adjust the number to what fits your screen
pd.set_option('display.max_columns', None) # Display any number of columns
# Convert the DataFrame (or a part of it) to HTML and display
display(HTML(df[['prompt', 'actual', 'predicted']].to_html()))
```
